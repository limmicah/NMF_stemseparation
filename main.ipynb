# **NMF**

```python
mix_stft = YouSteppedOutOfADream_saxophone_bass_C7_stft  # change to the mix you want to separate
mixture_signal = YouSteppedOutOfADream_saxophone_bass_C7  # time-domain signal used for length and phase reference
V = np.abs(mix_stft) + 1e-8  # shape: (n_freq, n_frames); add small eps to avoid exact zeros
n_components = 2  # adjust number of components (sources/patterns)
model = NMF(n_components=n_components, init='random', random_state=0, max_iter=1000)
W = model.fit_transform(V)  # shape: (n_freq, n_components)
H = model.components_
S = W.shape[1]
```

```python
# After NMF, each component can be expressed as a frequency mask over time
WH = W @ H + 1e-8  # reconstructed magnitude (avoid divide-by-zero)
# Create subplots sized to number of components
fig, axs = plt.subplots(nrows=1, ncols=S, figsize=(4*S, 4))
filtered_spectrograms = []
masks = []
for i in range(S):
    ax = axs if S == 1 else axs[i]
    ax.set_title(f"Frequency Mask of Component {i+1}")
    # component magnitude estimate (n_freq, n_frames)
    Si = (W[:, [i]] @ H[[i], :])
    # soft mask
    mask_i = Si / WH
    filtered = mask_i * V
    D = librosa.amplitude_to_db(filtered, ref=np.max)
    librosa.display.specshow(D, y_axis='hz', sr=sr, hop_length=HOP, x_axis='time', cmap=matplotlib.cm.jet, ax=ax)
    filtered_spectrograms.append(filtered)
    masks.append(mask_i)
```

**Reconstruction**

```python
# Reconstruction from filtered magnitude spectrograms using mixture phase
mixture_phase = np.angle(mix_stft)
reconstructed_sounds = []
for i in range(S):
    complex_spec = filtered_spectrograms[i] * np.exp(1j * mixture_phase)
    # invert to time domain; ensure length matches original signal
    y_i = librosa.istft(complex_spec, hop_length=HOP, length=len(mixture_signal))
    reconstructed_sounds.append(y_i)
# You can listen in a notebook: ipd.display(ipd.Audio(reconstructed_sounds[0], rate=sr))
```

```python
# Plotting the waveform
colors = ['r', 'g','b','c']
fig, axs = plt.subplots(nrows=S, ncols=1, sharex=True, figsize=(10, 8))
for i in range(S):
    ax = axs if S == 1 else axs[i]
    color = colors[i % len(colors)]
    librosa.display.waveshow(reconstructed_sounds[i], sr=sr, color=color, ax=ax, label=f'Source {i}', x_axis='time')
    ax.set(xlabel='Time [s]')
    ax.legend()
```
